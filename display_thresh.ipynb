```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

'''
Author: Georgie Nahas
Simpple functino to display some threholded images for display
'''
thresholds = [127, 133, 150, 170, 185, 200, 225]


def threshold_image(image, min_val=127, max_val=255):
    """
    input: image
    Optional: hyperpatameter pixel values for thresholding 
    return: count of white pixels normalized by image size
    """
    _, thresh = cv2.threshold(image, min_val, max_val, cv2.THRESH_BINARY)

    return thresh



orig_image = '/Users/georgienahass/Desktop/fall_classes_PhD/CS415/CXR_png.nosync/CHNCXR_0359_1.png'
image = cv2.imread(orig_image, cv2.IMREAD_GRAYSCALE)

for thresh in thresholds:
    new_img = threshold_image(image,  min_val=thresh)
# plt.imshow(image, cmap='gray')
# plt.axis(False)
# plt.show()

    plt.imshow(new_img,cmap='gray')
    plt.axis(False)
    plt.show()
```
