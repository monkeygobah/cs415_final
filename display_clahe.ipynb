```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


'''
Author: Georgie Nahas
Simpple function to display some CLAHE images for display
'''
def increase_contrast(image, box_size=8, lim = 2):
    '''
    Increase contrast of image
    Input: Image
    Optional: box size and clip limit- can finetune these
    Output: contrast increased image  
    '''
    clahe = cv2.createCLAHE(clipLimit=lim, tileGridSize=(box_size, box_size))
    return clahe.apply(image)


orig_image = '/Users/georgienahass/Desktop/fall_classes_PhD/CS415/CXR_png.nosync/CHNCXR_0359_1.png'
image = cv2.imread(orig_image, cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.axis(False)
plt.show()

cont_image = increase_contrast(image)
plt.imshow(cont_image,cmap='gray')
plt.axis(False)
plt.show()
```
